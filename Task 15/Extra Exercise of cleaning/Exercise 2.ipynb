### Step 1. Import the necessary libraries

```python
import pandas as pd 
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
df = pd.read_csv("u.user",sep="|")
df
```

### Step 3. Assign it to a variable called users.

```python
users =  pd.read_csv("u.user",sep="|")
users.columns
```

### Step 4. Discover what is the mean age per occupation

```python
mean_age = users.groupby('occupation')['age'].mean()

mean_age.reset_index()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python

gender_count = df.groupby('occupation')['gender'].value_counts().unstack()

gender_count.fillna(0, inplace=True)

gender_count['Male_Ratio'] = gender_count['M'] / (gender_count['M'] + gender_count['F'])


male_ratio_sorted = gender_count['Male_Ratio'].sort_values(ascending=False)

male_ratio_sorted
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python

age_stats = df.groupby('occupation')['age'].agg(['min', 'max']).reset_index()
age_stats
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python

mean_age_per_group = df.groupby(['occupation', 'gender'])['age'].mean().reset_index()


mean_age_per_group.sort_values(by=['occupation', 'gender'], inplace=True)


mean_age_per_group
```

### Step 8.  For each occupation present the percentage of women and men

```python

occupation_counts = df['occupation'].value_counts()


gender_counts = df.groupby(['occupation', 'gender']).size().unstack(fill_value=0)

gender_percentage = (gender_counts.div(occupation_counts, axis=0) * 100).reset_index()


gender_percentage.columns = ['occupation', 'Female_Percentage', 'Male_Percentage']

gender_percentage
```
